```python
import quadrotor
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = quadrotor.Model()
```

```python
def u(t):
    return np.array([0, 0, 0, np.sin(t)])

p = np.array([1, 1, 1, 1])

t = np.arange(0, 10, 0.01)
res = model.simulate(t, u)

plt.plot(res['t'], res['y'].T);
```
